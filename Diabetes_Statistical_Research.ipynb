```python
#Measure of dispersion
#SD, VARIANCE, IQR AND RANGE

groups=df.groupby('Diabetes_binary')

def dispersion_stats(col):
    return groups[col].agg([
        ('Mean', 'mean'),
        ('Std Dev','std'),
        ('Variance', 'var'),
        ('Min','min'),
        ('Max','max'),
        ('Range', lambda x: x.max() - x.min()),
        ('Q1', lambda x: x.quantile(0.25)),
        ('Q3', lambda x: x.quantile(0.75)),
        ('IQR', lambda x: x.quantile(0.75) - x.quantile(0.25))
    ])

#Apply to continuous data columns
bmi_stats = dispersion_stats('BMI')
menthlth_stats = dispersion_stats('MentHlth')
physhlth_stats = dispersion_stats('PhysHlth')

# Display
print("BMI Dispersion:\n", bmi_stats, "\n")
print("Mental Health Dispersion:\n", menthlth_stats, "\n")
print("Physical Health Dispersion:\n", physhlth_stats)
#Relatively High Std dev and Variance for people with diabetes
```

```python
#Bayes' Theoram
#“What is the probability that a person is diabetic given they reported Poor general health (GenHlth = 5)?”

total = len(df)

p_diabetes = (df['Diabetes_binary']==1).mean()

p_poor_health = (df['GenHlth'] == 5).mean()

p_poor_health_given_diabetes = df[df['Diabetes_binary']==1]['GenHlth'].value_counts(normalize=True).get(5, 0)

# Apply Bayes' Theorem
p_diabetes_given_poor_health = (p_poor_health_given_diabetes * p_diabetes) / p_poor_health

print(f"P(Diabetes=1 | GenHlth=5): {p_diabetes_given_poor_health:.4f}")
```

```python
#inferential statistics
#What’s the likely range for average BMI among these two groups in the population?
#CI estimation

diabetic_bmi = df[df['Diabetes_binary'] == 1]['BMI']

non_diabetic_bmi = df[df['Diabetes_binary'] ==0]['BMI']

import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin = std_err * stats.t.ppf((1 + confidence)/ 2, df = n-1)
    return mean-margin, mean+margin

# Confidence interval for diabetics
ci_diabetic = confidence_interval(diabetic_bmi)

# Confidence interval for non-diabetics
ci_nondiabetic = confidence_interval(non_diabetic_bmi)

print("95% Confidence Interval for mean BMI (Diabetics):", ci_diabetic)
print("95% Confidence Interval for mean BMI (Non-Diabetics):", ci_nondiabetic)
```

```python
#Those are clear and non-overlapping confidence intervals,
#which already suggests that diabetics have a statistically significantly higher BMI 
#than non-diabetics.
```

```python
#HYPOTHESIS TESTING

# “Is there a statistically significant difference in BMI between diabetics and non-diabetics?”
# “Does general health affect the likelihood of having diabetes?”
# “Do smokers have higher chances of diabetes compared to non-smokers?”

from scipy.stats import ttest_ind

#run t-test to check pvalues

t_stat, p_value = ttest_ind(diabetic_bmi,non_diabetic_bmi,equal_var=False)
print(f"T-statistic: {t_stat: .4f}")
print(f"P-value: {p_value: .4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis — significant difference in BMI.")
else:
    print("Fail to reject the null — no significant difference in BMI.")
```

```python
physHlth_Diabeties = df[df['Diabetes_binary'] == 1]['PhysHlth']
physHlth_non_diabeties = df[df['Diabetes_binary'] == 0]['PhysHlth']

ment_health_diabeties = df[df['Diabetes_binary'] == 1]['MentHlth']
ment_health_non_diabeties = df[df['Diabetes_binary'] == 0]['MentHlth']

# T-test for PhysHlth
t_stat_phys, p_val_phys = stats.ttest_ind(physHlth_Diabeties, physHlth_non_diabeties, equal_var=False)

# T-test for MentHlth
t_stat_ment, p_val_ment = stats.ttest_ind(ment_health_diabeties, ment_health_non_diabeties, equal_var=False)

# Print results
print("T-test Results (PhysHlth):")
print("T-statistic:", round(t_stat_phys, 4))
print("P-value:", p_val_phys)

print("\nT-test Results (MentHlth):")
print("T-statistic:", round(t_stat_ment, 4))
print("P-value:", p_val_ment)
```

```python
# Chi-square-test(if two categorical variables are dependent on each other)
# Diabetes_binary (0 = No, 1 = Yes) against each of these:
# 	•	HighBP
# 	•	HighChol
# 	•	Smoker
# 	•	Stroke
# 	•	Fruits
# 	•	Veggies
from scipy.stats import chi2_contingency

contigency_table = pd.crosstab(df['Diabetes_binary'], df['Smoker'])

chi2, p , dof, expected = chi2_contingency(contigency_table)
print("Chi2 Statistic:", chi2)
print("Degrees of Freedom:", dof)
print(f"P-Value:{p : .5f}")
```

```python
features = ['HighBP', 'HighChol', 'Smoker', 'Stroke', 'Fruits', 'Veggies']

for feature in features:
    print(f"Chi-Square Test: Diabetic_binary vs {feature}")
    table = pd.crosstab(df['Diabetes_binary'],df[feature])

    chi2, p, dof, expected = chi2_contingency(table)
     # Display results
    print(f"Chi2 Statistic = {chi2:.4f}")
    print(f"Degrees of Freedom = {dof}")
    print(f"P-Value = {p:.4f}")
    
    # Interpretation
    if p < 0.05:
        print(" Statistically significant association (Reject Null Hypothesis)")
    else:
        print(" No significant association (Fail to Reject Null Hypothesis)")
 
```

```python
# The higher the Chi2 statistic, the stronger the deviation from expected independence → suggesting a stronger association.
```

```python
#In our case target variable is binary so we will proceed with
# Logistic Regression 
import statsmodels.api as sm

features = ['HighBP','HighChol','BMI','Smoker','Stroke','Fruits','Veggies']

X = df[features]
y= df['Diabetes_binary']

X = sm.add_constant(X)

model = sm.Logit(y, X).fit()

print(model.summary())
```

```python
#PREDICTIVE MODELLING

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

features = ['HighBP', 'HighChol', 'BMI', 'Smoker', 'Stroke', 'Fruits', 'Veggies']

X= df[features]
y= df['Diabetes_binary']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=30, stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=30)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Accuracy
print("XG BOOST MODEL Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
